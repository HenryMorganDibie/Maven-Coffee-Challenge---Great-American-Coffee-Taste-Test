```python
import pandas as pd

# Load the datasets
questions = pd.read_csv("QUESTION_KEY.csv")
results = pd.read_csv("GACTT_RESULTS_ANONYMIZED_v2.csv")

# Target Audience Analysis
age_preferences = results['What is your age?'].value_counts()
print("Age Group Preferences:\n", age_preferences)

# Brewing Preferences
brewing_preferences = results['How do you brew coffee at home?'].str.split(', ').explode().value_counts()
print("\nBrewing Preferences:\n", brewing_preferences)

# Additives Preferences
additives_preferences = results['Do you usually add anything to your coffee?'].str.split(', ').explode().value_counts()
print("\nAdditives Preferences:\n", additives_preferences)

# Product Offering - Coffee Bean Preferences
coffee_bean_preferences = results.filter(like='Coffee').filter(like='Personal Preference').stack().value_counts()
print("\nCoffee Bean Preferences:\n", coffee_bean_preferences)

# Product Offering - Brewing Equipment Preferences
equipment_preferences = results.filter(like='How do you brew coffee at home?').stack().value_counts()
print("\nBrewing Equipment Preferences:\n", equipment_preferences)

# Product Offering - Favorite Overall Coffee
favorite_coffee_preferences = results['Lastly, what was your favorite overall coffee?'].value_counts()
print("\nFavorite Overall Coffee Preferences:\n", favorite_coffee_preferences)

# Pricing Strategy - Maximum Price Perception
max_price_perception = results['What is the most you\'d ever be willing to pay for a cup of coffee?'].value_counts()
print("\nMaximum Price Perception:\n", max_price_perception)

# Pricing Strategy - Value Perception at Cafes
value_perception_at_cafes = results['Do you feel like you’re getting good value for your money when you buy coffee at a cafe?'].value_counts()
print("\nValue Perception at Cafes:\n", value_perception_at_cafes)
```

```python
# Extract and analyze spending patterns
spending_patterns = results['Approximately how much have you spent on coffee equipment in the past 5 years?'].str.split(', ').explode().value_counts()

# Print the results
print("\nSpending Patterns:\n", spending_patterns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
questions = pd.read_csv("QUESTION_KEY.csv")
results = pd.read_csv("GACTT_RESULTS_ANONYMIZED_v2.csv")

# Target Audience Analysis
age_preferences = results['What is your age?'].value_counts()
print("Age Group Preferences:\n", age_preferences)

# Visualization for Age Group Preferences
plt.figure(figsize=(10, 6))
sns.barplot(x=age_preferences.index, y=age_preferences.values, palette='viridis')
plt.title('Age Group Preferences')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

# Brewing Preferences
brewing_preferences = results['How do you brew coffee at home?'].str.split(', ').explode().value_counts()
print("\nBrewing Preferences:\n", brewing_preferences)

# Visualization for Brewing Preferences
plt.figure(figsize=(12, 6))
sns.barplot(x=brewing_preferences.index, y=brewing_preferences.values, palette='Set2')
plt.title('Brewing Preferences')
plt.xlabel('Brewing Method')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Additives Preferences
additives_preferences = results['Do you usually add anything to your coffee?'].str.split(', ').explode().value_counts()
print("\nAdditives Preferences:\n", additives_preferences)

# Visualization for Additives Preferences
plt.figure(figsize=(12, 6))
sns.barplot(x=additives_preferences.index, y=additives_preferences.values, palette='pastel')
plt.title('Additives Preferences')
plt.xlabel('Additive')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Product Offering - Coffee Bean Preferences
coffee_bean_preferences = results.filter(like='Coffee').filter(like='Personal Preference').stack().value_counts()
print("\nCoffee Bean Preferences:\n", coffee_bean_preferences)

# Visualization for Coffee Bean Preferences
plt.figure(figsize=(8, 6))
sns.countplot(x=coffee_bean_preferences, palette='Blues')
plt.title('Coffee Bean Preferences')
plt.xlabel('Preference Rating')
plt.ylabel('Count')
plt.show()

# Pricing Strategy - Maximum Price Perception
max_price_perception = results['What is the most you\'d ever be willing to pay for a cup of coffee?'].value_counts()
print("\nMaximum Price Perception:\n", max_price_perception)

# Visualization for Maximum Price Perception
plt.figure(figsize=(10, 6))
sns.barplot(x=max_price_perception.index, y=max_price_perception.values, palette='coolwarm')
plt.title('Maximum Price Perception')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Pricing Strategy - Value Perception at Cafes
value_perception_at_cafes = results['Do you feel like you’re getting good value for your money when you buy coffee at a cafe?'].value_counts()
print("\nValue Perception at Cafes:\n", value_perception_at_cafes)

# Visualization for Value Perception at Cafes
plt.figure(figsize=(6, 6))
sns.countplot(x=value_perception_at_cafes, palette='Set1')
plt.title('Value Perception at Cafes')
plt.xlabel('Value Perception')
plt.ylabel('Count')
plt.show()
```

```python
# Extract and analyze spending patterns
spending_patterns = results['Approximately how much have you spent on coffee equipment in the past 5 years?'].str.split(', ').explode().value_counts()

# Print the results
print("\nSpending Patterns:\n", spending_patterns)

# Visualization for Spending Patterns
plt.figure(figsize=(12, 6))
sns.barplot(x=spending_patterns.index, y=spending_patterns.values, palette='viridis')
plt.title('Spending Patterns on Coffee Equipment in the Past 5 Years')
plt.xlabel('Spending Range')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```
